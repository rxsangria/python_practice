### Setting up environment

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

### Allow visualizations to be shown inline 

```python
%matplotlib inline
```

### Simple example

```python
x = np.linspace(0,5,11)
y = x**2
```

```python
x
```

```python
y
```

```python
#FUNCTIONAL METHOD
plt.plot(x,y)
# plt.show is printing out the plot,necessary if not using jupyter notebook
# plt.show()
#label the axes
plt.xlabel('X Label')
plt.ylabel('Y Label')
#Create a title
plt.title('Title')
```

### Create multiplots on the same canvas

```python
#subplot(number of rows, number of columns, plot number)
plt.subplot(1,2,1)
# 'r' means red, 'b' means blue
plt.plot(x,y,'r')

plt.subplot(1,2,2)
plt.plot(x,y,'b')
```

### Object oriented method - more in depth but have better control

```python
#blank canvas
fig = plt.figure()
#add set of axes arguments [left, bottom, width, height] 
axes = fig.add_axes([0.1,0.1,0.8,0.8])
#plot on the axes
axes.plot(x,y)
#label the axes, title
axes.set_xlabel('X Label')
axes.set_ylabel('Y Label')
axes.set_title('Title')
```

```python
#creating a plot within a plot - like a zoom in
fig = plt.figure()
#all axes parameters must be between 0 and 1. 0.2 means 20% of canvas
axes1 = fig.add_axes([0.1,0.1,0.8,0.8])
axes2 = fig.add_axes([0.15,0.55,0.4,0.3])
# parameters are 'how far from the left, how far up from the bottom, how wide, how tall'
axes1.plot(x,y)
axes1.set_title('Big Plot')
axes2.plot(y,x)
axes2.set_title('Small Plot')
```

### Create subplots using object oriented method

```python
fig,axes = plt.subplots(nrows=3,ncols=3) # subplots(nrows,ncols) allows you to add more plots to the 'grid'
plt.tight_layout()
#specify which plot to write by indexing as shown 
axes[0][1].plot(x,y,'r--')
axes[0][1].set_title('XY plot INDEX [0][1]')
axes[0][2].plot(y,x,'g*')
axes[0][2].set_title('YX plot INDEX [0][2]')
axes[0][0].set_ylabel('INDEX [0] ROW')
axes[1][0].set_ylabel('INDEX [1] ROW')
axes[2][0].set_ylabel('INDEX [2] ROW')
axes[2][0].set_xlabel('INDEX [0] COLUMN')
axes[2][1].set_xlabel('INDEX [1] COLUMN')
axes[2][2].set_xlabel('INDEX [2] COLUMN')
```

### Figure size and dpi

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y,'r*',label='xy')
ax.plot(y,x,label='yx')

ax.legend(loc=5)
```

```python
fig.savefig('my_picture.png',dpi=200)
```

```python
fig = plt.figure(figsize=(5,4))
ax = fig.add_axes([0,0,1,1])
#changing the markers, line color, line width, style, size, etc. 
ax.plot(x,y,color='#A37EFB',lw=2,ls='-.',marker='o',markersize=7,markerfacecolor='red',markeredgewidth=2,markeredgecolor='green') #HEX CODE
```

### control over axis appearance

```python
fig = plt.figure()
ax=fig.add_axes([0,0,1,1])
ax.plot(x,y)
#zooming in by setting x and y axis limits
ax.set_xlim([0,1]) #set the limit of the view starting at 0 and ending at 1 for the x axis
ax.set_ylim([0,2]) #set the limit of the view starting at 0 and ending at 2 for the y axis
```
